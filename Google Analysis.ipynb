Here I will be analyzing the Google search trends on the queries based on “Machine Learning”, so let’s create a DataFrame of the top 10 countries which search for “Machine Learning” on Google:

```python
trends.build_payload(kw_list=["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by="Machine Learning", ascending=False)
data = data.head(10)
print(data)
```

So, according to the above results, the search queries based on “Machine learning” are mostly done in China. We can also visualize this data using a bar chart:

```python
trends.build_payload(kw_list=["Chinese Balloon"])
data = trends.interest_by_region()
data = data.sort_values(by="Chinese Balloon", ascending=False)
data = data.head(10)
print(data)
```

```python
trends.build_payload(kw_list=["Super Bowl"])
data = trends.interest_by_region()
data = data.sort_values(by="Super Bowl", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName", 
                        y="Super Bowl", 
                        figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Johnny Deep'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Johnny Deep'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Johnny Deep', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
